```python
def fibonacci(n):
    """Return a list of the first n Fibonacci numbers."""
    fib = []
    a, b = 0, 1
    while len(fib) < n:
        fib.append(b)
        a, b = b, a + b
    return fib

```

```python
import random

random.seed(str("rob-t-1"))

fibonacci_list = fibonacci(50)
fibonacci_list = random.sample(fibonacci_list, 5)

random_list = random.sample(range(max(fibonacci_list)), 9995)

new_list = random_list + fibonacci_list

# shuffle new list
random.shuffle(new_list)

# convert new list to string and save to file
```

```python
",".join(str(x) for x in new_list)
```

```python
# Find all the fibonacci numbers in new_list and sum them
sum = 0
for i in new_list:
    if i in fibonacci_list:
        sum += i

print(sum)  
```

```python
import math

n = 20
k = 5

combinations = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

print(combinations)
```
